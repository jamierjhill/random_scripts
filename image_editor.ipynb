```python
!pip install pillow


from PIL import Image, ImageEnhance, ImageFilter
import os

path = '/content/drive/MyDrive/Colab Notebooks/image_editor/images'
pathOut = '/content/drive/MyDrive/Colab Notebooks/image_editor/edited_images'

for filename in os.listdir(path):
    img = Image.open(f"{path}/{filename}")

    edit = img.filter(ImageFilter.SHARPEN).convert('L')
    edit = ImageEnhance.Contrast(edit).enhance(1.5)
    edit = ImageEnhance.Brightness(edit).enhance(1.5)
    edit = ImageEnhance.Sharpness(edit).enhance(2)

    clean_name = os.path.splitext(filename)[0]

    edit.save(f"{pathOut}/{clean_name}_edited.jpg")

    print("Images are edited and in edited_images folder")

```
